Predict sales depending upon the adv bug spend upon TV,news paper and radio

```python
import numpy as np                                # numpy is imported
import pandas as pd                               # pandas is imported
import matplotlib.pyplot as plt                   # matplotlib is imported
import seaborn as sns                             # seaborn is imported

import warnings                                   # warnings is imported 
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ingledarshan/BoardInfinity/main/Sales.csv")            # reading dataset
df
```

```python
df.drop('Unnamed: 0', axis=1,inplace=True)
df
# removing unwanted columns & using inplace=True for saving changes in original dataset
```

```python
X=df.iloc[:,:-1] # selected all columns from dataframe except last one
X
```

```python
y=df.iloc[:,-1] ## selected only lastcolumn from the dataframe
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=['TV', 'radio', 'newspaper'])
X_train
```

```python
X_test = pd.DataFrame(X_test, columns=['TV', 'radio', 'newspaper'])
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Evaluation

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
